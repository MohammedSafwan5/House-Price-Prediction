## House Price Prediction Using Linear Regression

### Housing Price Prediction Based on Housing Area

```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv("C:\ML\Datasets\Housing.csv")
#We have to perform Linear Regression based on Housing Area. So we have to separate the area and price column.
data = dataset.iloc[:, :2]
data.head()
```

#### Visualize the Data

```python
plt.xlabel("Area")
plt.ylabel("Price")
plt.title("Housing Price vs Area plot")
plt.scatter(data['area'], data.price)
```

```python
# Area is the independent variable.
# .values takes only the feature values, not the column titles.
X = data.iloc[:, 1].values.reshape((-1,1))
# X = data.drop('price', axis = 'columns')

# Housing price is the dependent variable.
y = data.iloc[:, 0].values
# y = data.price

print("Housing area : ", X)
print("Housing Price : ", y)
```

```python
# If we are using .values.reshape((-1,1))
X = np.array(X).reshape((-1,1))
y = np.array(y)
```

### Training the Model

```python
model = LinearRegression()
model.fit(X,y)
```

### Predicting Price 

```python
x = int(input("Enter the area of the house : "))
LandArea = [[x]]
PredictedResult = model.predict(LandArea)
print(PredictedResult)
```

### Theory Calculation

```python
m = model.coef_
c = model.intercept_
print("The slope of the model is : ", m)
print("The intercept of the model is : ", c)
```

```python
result = m*x + c
print("The price from the theory is : ", result)
```

```python
if(PredictedResult == result):
    print("The predicted and theoretical result is equal.")
else :
    print("The difference between predicted and theoretical result is : ", abs(PredictedResult - result))
```
